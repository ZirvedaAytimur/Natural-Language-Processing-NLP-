```python
import spacy
import nltk
from trnlp import *
```

```python
nlp = spacy.load('en_core_web_sm')
nlp.max_length = 1088508
```

```python
with open('tur_news_2005_10K-sentences.txt', encoding="utf8") as f:
    doc = nlp(f.read())
```

```python
len(doc)
```

```python
for token in doc[:2000]:
    print(f'{token.text:{30}} {token.pos_:{15}} {spacy.explain(token.pos_):{15}} {token.tag_:{10}} {spacy.explain(token.tag_)}')
```

```python
f = open('tur_news_2005_10K-sentences.txt', 'r', encoding='utf8')
doc = f.read()
```

```python
text = nltk.word_tokenize(doc[:2000])
```

```python
nltk.pos_tag(text)
```

```python
obj = TrnlpWord()
```

```python
for token in simple_token(doc[:2000]):
    obj.setword(token)
    print(obj)
```
