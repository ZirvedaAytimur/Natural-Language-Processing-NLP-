```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# prepare data for training our doc2vec model
data = ["I love machine learning. Its awesome.",
        "I love coding in python",
        "I love building chatbots",
        "they chat amagingly well",
        "You are awesome.",
        "This was the best day of my life.",
        "Going to school is very tiring.",
        "Her project is about to finish.",
        "I will always remember you."]

tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), 
                              tags=[str(i)]) for i, _d in enumerate(data)]
tagged_data
```

```python
#number of iterations
max_epochs = 200
#Dimensionality of the feature vectors.
vec_size = 40
#The initial learning rate.
alpha = 0.025

model = Doc2Vec(size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                #Ignores all words with total frequency lower than this.
                min_count=1,
                #Defines the training algorithm. If dm=1, ‘distributed memory’ (PV-DM) is used. Otherwise, distributed bag of words (PV-DBOW) is employed.
                dm =1)

model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha
```

```python
model= Doc2Vec.load("d2v.model")
```

```python
test_data = sent_tokenize("I love you.".lower())
v1 = model.infer_vector(test_data)
```

```python
similar_doc = model.docvecs.most_similar(positive=[v1])
print(similar_doc)
print(data[int(similar_doc[0][0])])
```
