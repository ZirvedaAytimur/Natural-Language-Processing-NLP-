```python
from nltk.tokenize import sent_tokenize, word_tokenize
from collections import Counter
```

```python
from os.path import join
from typing import List

from jpype import JClass, JString, getDefaultJVMPath, shutdownJVM, startJVM
```

```python
ZEMBEREK_PATH: str = join('..', '..', 'bin', 'zemberek-full.jar')

startJVM(
    getDefaultJVMPath(),
    '-ea',
    f'-Djava.class.path={ZEMBEREK_PATH}',
    convertStrings=False
)

TurkishMorphology: JClass = JClass('zemberek.morphology.TurkishMorphology')
TurkishSpellChecker: JClass = JClass(
    'zemberek.normalization.TurkishSpellChecker'
)

morphology: TurkishMorphology = TurkishMorphology.createWithDefaults()

spell_checker: TurkishSpellChecker = TurkishSpellChecker(morphology)
```

```python
def getwords(text):
    return (w for sent in sent_tokenize(text) for w in word_tokenize(sent))
```

```python
def normalize(text): # remove non-ascii, convert to lowercase
    return text.encode('ascii', 'ignore').lower().decode()
```

```python
f = open('tur_mixed_2014_10K-sentences.txt', 'r', encoding='utf8')
doc = f.read()
```

```python
count = Counter(getwords(normalize(doc)))
high_frequency = [w for w, c in count.items() if c > 5]
low_frequency = [w for w, c in count.items() if c < 5]
```

```python
high_frequency
```

```python
for i, word in enumerate(low_frequency):
        if spell_checker.suggestForWord(JString(word)):
            if not spell_checker.check(JString(word)):
                spellCheckedWord = str(spell_checker.suggestForWord(JString(word))[0])
                print(f"{word:{20}} {spellCheckedWord:{30}}")
```

```python
shutdownJVM()
```
