```python
import re
import matplotlib.pyplot as plt
import pandas as pd
import nltk
```

```python
f = open('hurriyet.txt', 'r', encoding='utf8')
text = f.read()

words = re.sub(r'[^\w\s]', '', text).split()
```

```python
len(pd.Series(nltk.ngrams(words, 2)).value_counts())
```

```python
count_bigrams = (pd.Series(nltk.ngrams(words, 2)).value_counts())
len(count_bigrams[count_bigrams > 2000])
```

```python
len(pd.Series(nltk.ngrams(words, 3)).value_counts())
```

```python
count_trigrams = (pd.Series(nltk.ngrams(words, 3)).value_counts())
len(count_trigrams[count_trigrams > 300])
```

```python
bigrams_series = (count_bigrams)[:20]
trigrams_series = (count_trigrams)[:20]
```

```python
bigrams_series.sort_values().plot.barh(color='purple', width=.9, figsize=(12, 8))
plt.title('20 Most Frequently Occuring Bigrams')
plt.ylabel('Bigram')
plt.xlabel('# of Occurances')
```

```python
trigrams_series.sort_values().plot.barh(color='pink', width=.9, figsize=(12, 8))
plt.title('20 Most Frequently Occuring Triagram')
plt.ylabel('Trigram')
plt.xlabel('# of Occurances')
```
